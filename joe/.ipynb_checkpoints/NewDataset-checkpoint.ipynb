```python
# Using the GREYC keystroke dataset from http://www.vmonaco.com/keystroke-datasets
import pandas as pd;
import numpy as np;
import os;
```

```python
genuine_list=[];
imposter_list=[];

for root, dirs, files in os.walk("data/user_001", topdown=False):
    for subdir in dirs:
        try:
            if "genuine" in root:
                genuine_list.append(open(os.path.join(root,subdir)+"/l_raw_press.txt").readlines())
            else:
                imposter_list.append(open(os.path.join(root,subdir)+"/l_raw_press.txt").readlines())
#         ignore files that aren't timestampped keystroke data
        except: 
            pass;
```
